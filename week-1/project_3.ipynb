print("ANNUITY PLAN CALCULATOR")

# declare variables and get input from the user

p = float(input("Enter the principal amount>>> "))
m = float(input("How much do you deposit periodically>>> "))
r = float(input("Enter the annual interest rate>>> "))
n = float(input("Enter the number of times the interest is compounded per year>>> "))
t = float(input("Enter the number of years>>> "))

# calculate the annuity plan
a = p * m * t * (((1 + r / n) ** (n * t) - 1) / (r / n))

# print the output
print(f"The Annuity Plan is {a:.2f}")
