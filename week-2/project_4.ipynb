```python
#experience and age ATR program

# create variables which takes input of experience and age
exp  = int(input("Kindly enter your years of experience>>>"))
age = int(input("Kindly enter your age in years>>>"))

#create conditional statements to determine ATR

if exp > 25 and age >= 55:
    print("Your Annual Tax Revenue(ATR) is N5,600,000")


elif exp > 20 and age >= 45:
    print("Your Annual Tax Revenue(ATR) is N4,480,000")


elif exp > 10 and age >= 35:
    print("Your Annual Tax Revenue(ATR) is N1,500,000")

elif exp < 10 and age < 35:
    print("Your Annual Tax Revenue(ATR) is N550, 000")

else:
    print("Invalid input. Please enter valid experience and age.")
```
